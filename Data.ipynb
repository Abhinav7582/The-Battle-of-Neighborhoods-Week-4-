# Foursquare API Data:

We will need data about different venues in different neighborhoods of that specific borough. In order to gain that information we will use "Foursquare" locational information. Foursquare is a location data provider with information about all the venues and events within an area of interest. Such information includes venue names, locations, photos. As such, the foursquare location platform will be used as the sole data source since all the stated required information can be obtained through the API.
After finding the list of neighborhoods, we then connect to the Foursquare API to gather information about venues inside each and every neighborhood. For each neighborhood, we have chosen the radius to be 100 meter.

The data retrieved from Foursquare contained information of venues within a specified distance of the venue and the postal codes. The information obtained of the venues are:

1.Office
2.Taluk
3.District
4.State
5.Pincode
6.Venue name
7.Venue pincode
8.Venue area
